```python
import pandas as pd
from sklearn.preprocessing import normalize

df= pd.read_csv("calyx_bubble.csv", sep=",")
data=df.set_index(['carbon', 'molecularFormula'])
```

```python
data.head()
```

```python

trans=data.transpose()
#trans=normalize(trans)
trans.head()
```

```python
data_scaled=normalize(trans)
data_scaled=pd.DataFrame(data_scaled, columns=trans.columns, index=trans.index)
data_scaled
```

```python
#data_scaled.to_csv("calyx_bubble.tsv",sep="\t")
```

```python
filter_df=data_scaled.transpose()
filter_df
```

```python
#get hydroxy nanns
filter_df=filter_df.reset_index()
OH=filter_df[filter_df['molecularFormula'].str.contains('O2')]
OH=OH[~OH['molecularFormula'].str.contains('N3')]
OH=OH.set_index(['carbon'])
OH=OH.drop(['molecularFormula'], axis=1)

#OH=filter_df.filter(like='O2', axis=0)
OH
```

```python

#get non hydroxy nanns

n=filter_df[filter_df['molecularFormula'].str.endswith('O')]
n=n[~n['molecularFormula'].str.contains('N3')]
n=n.set_index(['carbon'])
n=n.drop(['molecularFormula'], axis=1)
n
```

```python
calyx_sum_n=n.groupby(level=0).sum().reset_index()
calyx_sum_n
```

```python
calyx_sum_OH=OH.groupby(level=0).sum().reset_index()
calyx_sum_OH
```

```python
#get exudate data

df= pd.read_csv("exu_bubble.csv", sep=",")
data=df.set_index(['carbon', 'molecularFormula'])
trans=data.transpose()
#trans=normalize(trans)
#trans.head()
data_scaled=normalize(trans)
data_scaled=pd.DataFrame(data_scaled, columns=trans.columns, index=trans.index)
data_scaled
filter_df=data_scaled.transpose()
#filter_df
#get hydroxy nanns
filter_df=filter_df.reset_index()
OH=filter_df[filter_df['molecularFormula'].str.contains('O2')]
OH=OH[~OH['molecularFormula'].str.contains('N3')]
OH=OH.set_index(['carbon'])
OH=OH.drop(['molecularFormula'], axis=1)

#OH=filter_df.filter(like='O2', axis=0)
#OH

#get non hydroxy nanns

n=filter_df[filter_df['molecularFormula'].str.endswith('O')]
n=n[~n['molecularFormula'].str.contains('N3')]
n=n.set_index(['carbon'])
n=n.drop(['molecularFormula'], axis=1)
```

```python
exu_sum_n=n.groupby(level=0).sum().reset_index()
exu_sum_n
```
