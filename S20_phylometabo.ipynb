```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import squareform
from scipy.spatial.distance import pdist

df= pd.read_csv("/Users/delser/Git/phd/input_pairwise_ntw.txt",index_col='cluster', sep="\t")
```

```python
df=df.transpose()
header=df.columns.values.tolist()
df
```
